```python
import matplotlib.pyplot as plt
import sklearn
import numpy as np
import pandas as pd


df = pd.read_csv("covid19.csv", sep = '\t' )
#df["total"] = df.groupby["state"]
df.head()
```

Grpah to show number of deaths from corona each day


```python
plt.figure(figsize=(15,10), dpi=100)
df["date1"] =df["date"].str[:5]
#df.sort_values('date1')
plt.title("Number of deaths from corona each day")
plt.plot(df["date1"], df["deaths"])
plt.xlabel("date in year 2020")
plt.xticks(rotation=90, size = 8)
plt.ylabel("Number of deaths")
plt.show()
```

```python
plt.figure(figsize=(15,10), dpi=100)
keys = [state for state, df in df.groupby(['state'])]
plt.bar(keys,df.groupby(['state']).sum()['deaths'])
plt.title("Number of deaths from corona in every state")
plt.ylabel('Deaths in each state')
plt.xlabel('state')
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

Number of deaths in each state(sorted)

```python
df1 = df.groupby(['state']).sum()
df1.sort_values('deaths')
final_df = df1.sort_values(by=['deaths'], ascending=False)
final_df
```

```python

fig, ax1 = plt.subplots(constrained_layout=True)


ax2 = ax1.twinx()
ax1.plot(df["state"],df['deaths'], 'r^--')
ax2.plot(df["state"], df['cases'],'y.-')
plt.title("Number of Cases and Deaths in each state")
ax1.set_xlabel('State')
ax1.set_ylabel('Deaths', color='g')
ax2.set_ylabel('Cases', color='b')

ax1.set_xticklabels(df['state'], rotation='vertical', size=8)
plt.show()
```
